```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python

col_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(r'ratings.csv', names=col_names)
```

```python
df.head()
```

```python
movies_names_genre=['item_id','movies', 'genre']
af= pd.read_csv(r'movies.csv', names= movies_names_genre)
```

```python
af.head()
```

```python
df= pd.merge(df, af, on='item_id')
df.head()
```

```python
# Visualization imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('movies')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('movies')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('movies')['rating'].mean())
ratings.head()
```

```python
ratings['noofratings'] = pd.DataFrame(df.groupby('movies')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
plt.figure(figsize=(10,8))
ratings['noofratings'].hist(bins=60)
```

```python
sns.jointplot(x='rating',y='noofratings' , data=ratings,alpha=0.5)
```

```python
samovies = df.pivot_table(index='user_id',columns='movies', values='rating')
samovies.head()
```

```python
# rated movie
ratings.sort_values('noofratings',ascending=False).head(500)
```

```python
# let takes two movies
Transformers_user_ratings = samovies['Transformers (2007)']
Fountain_user_ratings = samovies['Fountain, The (2006)']
Transformers_user_ratings.head()
```

```python
# use corrwith( method)
same_to_Transformers = samovies.corrwith(Transformers_user_ratings)
same_to_Fountain = samovies.corrwith(Fountain_user_ratings)
```

```python
corr_Transformers = pd.DataFrame(same_to_Transformers,columns=['Correlation'])
corr_Transformers.dropna(inplace=True)
corr_Transformers.head()
```

```python
corr_Transformers.sort_values('Correlation',ascending=False).head(20)
```

```python
# less than 200 reviews filtering out movies
corr_Transformers = corr_Transformers.join(ratings['noofratings'])
corr_Transformers.head()
```

```python
corr_Transformers[corr_Transformers['noofratings']>10].sort_values('Correlation',ascending=False).head()
```

```python
corr_Fountain = pd.DataFrame(same_to_Fountain, columns=['Correlation'])
corr_Fountain.dropna(inplace=True)
corr_Fountain = corr_Fountain.join(ratings['noofratings'])
corr_Fountain[corr_Fountain['noofratings']>10].sort_values('Correlation',ascending=False).head()
```
